# Task 2: Data Exploration

## Task 2.1 

```python
import matplotlib.pyplot as plt
```

```python
mask_2014 = copy_rainfall['Year'] == 2014
mask_2014
```

```python
year_2014 = copy_rainfall[mask_2014]
year_2014
```

```python
new_df_2014 = year_2014.pivot(index='Day', columns='Month', values='Rainfall amount (millimetres)')
new_df_2014
```

```python
highest_daily_rainfall = new_df_2014.max()
highest_daily_rainfall
```

```python
highest_daily_rainfall.plot(kind='bar')
plt.title('Highest Daily Rainfall of Each Month in 2014')
plt.ylabel('Rainfall amount (millimetres)')
plt.xlabel('Month')
plt.show()
```

## Task 2.2 

```python
mask_1517 =(copy_rainfall['Year'] >= 2015)&(copy_rainfall['Year'] <= 2017)
```

```python
rainfall_1517 = copy_rainfall[mask_1517]
rainfall_1517
```

```python
yearly_totals_1517 = rainfall_1517.groupby('Year')['Rainfall amount (millimetres)'].sum()
```

```python
yearly_totals_1517.plot(kind='bar')
plt.title('Total Rainfall by Year (2015-2017)')
plt.ylabel('Rainfall amount (millimetres)')
plt.xlabel('Year')
plt.show()
```

```python
monthly_totals = rainfall_1517.groupby(['Year', 'Month'])['Rainfall amount (millimetres)'].sum()
monthly_totals.plot(kind='bar')
plt.title('Monthly Rainfall Totals by Year (2015-2017)')
plt.ylabel('Rainfall amount (millimetres)')
plt.xlabel('Month')
plt.show()
```

```python
monthly_totals
```

```python
monthly_comparison = rainfall_1517.groupby(['Year', 'Month'])['Rainfall amount (millimetres)'].sum().unstack(0)
monthly_comparison
```

```python
monthly_comparison.plot(kind='bar', figsize=(20, 10))
plt.title('Monthly Rainfall Comparison by Year (2015-2017)')
plt.xlabel('Month')
plt.ylabel('Total Rainfall (millimetres)')
plt.legend(title='Year')
plt.show()
```

## Task 2.3 

```python
total_rainfall_by_year = copy_rainfall.groupby('Year')['Rainfall amount (millimetres)'].sum()
```

```python
highest_top3_years = total_rainfall_by_year.nlargest(3)
highest_top3_years
```

```python
lowest_top3_years = total_rainfall_by_year.nsmallest(3)
lowest_top3_years
```

```python
comparision_top3 = pd.DataFrame({
    'Highest': highest_top3_years,
    'Lowest': lowest_top3_years
})
comparision_top3
```

```python
comparision_top3.plot(kind='bar') 
plt.title('Comparison of Top3 Highest and Lowest Rainfall Years')
plt.ylabel('Total Rainfall amount (millimetres)')
plt.show()
```

```python
monthly_rainfall = copy_rainfall.groupby(['Year', 'Month'])['Rainfall amount (millimetres)'].sum().reset_index()
monthly_rainfall
```

```python
max_rainfall_months = monthly_rainfall.groupby(['Year'])['Rainfall amount (millimetres)'].max()
max_rainfall_months
```

```python
min_rainfall_months = monthly_rainfall.groupby(['Year'])['Rainfall amount (millimetres)'].min()
min_rainfall_months
```

```python
comparision_monthly_max_min = pd.DataFrame({
    'Highest': max_rainfall_months,
    'Lowest': min_rainfall_months
})
```

```python
comparision_monthly_max_min.plot(kind='bar',figsize=(15, 10)) 
plt.title('Comparison of Highest and Lowest Monthly Rainfall in ABC City (Last 10 Years)')
plt.ylabel('Monthly Rainfall amount (millimetres)')
plt.show()
```

```python
monthly_comparison_10_years = monthly_rainfall.groupby(['Year', 'Month'])['Rainfall amount (millimetres)'].sum().unstack(0)
monthly_comparison_10_years
```

```python
data_array = monthly_comparison_10_years.values
fig, ax = plt.subplots(figsize=(10, 8)) 
cax = ax.matshow(data_array, cmap='viridis')
cbar = fig.colorbar(cax)
ax.set_xticks(np.arange(len(monthly_comparison_10_years.columns)))
ax.set_xticklabels(monthly_comparison_10_years)
ax.set_yticks(np.arange(len(monthly_comparison_10_years.index)))
ax.set_yticklabels(monthly_comparison_10_years.index)
ax.set_xlabel('Year')
ax.set_ylabel('Month')
ax.set_title('Heatmap of Monthly Rainfall (2013-2023)')
cbar.set_label('Rainfall amount (millimetres)', rotation=270, labelpad=15)
for i in range(len(monthly_comparison_10_years.index)):
    for j in range(len(monthly_comparison_10_years.columns)):
        ax.text(j, i, '{:.1f}'.format(data_array[i, j]), ha='center', va='center', color='white')
plt.show()
```

```python
monthly_comparison_10_years.plot(kind='box')
plt.title('Monthly Rainfall Comparision in ABC City (Last 10 Years)')
plt.ylabel('Monthly Rainfall amount (millimetres)')
plt.xlabel('Year')
plt.show()
```
